```python
import numpy as np
import pandas as pd

# Assuming 'df' is the DataFrame containing your data and has been cleaned

# Convert 'Churn' to numeric (1 for 'Yes', 0 for 'No')
df['Churn'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

# Initialize the transition matrix
transition_matrix = np.zeros((2, 2))

# Calculate the transition counts based on consecutive churn statuses
for i in range(len(df) - 1):
    current_state = df['Churn'].iloc[i]
    next_state = df['Churn'].iloc[i + 1]
    transition_matrix[current_state, next_state] += 1

# Calculate the transition probabilities by normalizing the matrix
transition_probabilities = transition_matrix / transition_matrix.sum(axis=1, keepdims=True)

# Display the transition matrix and transition probabilities
print("Transition Matrix:")
print(transition_matrix)
print("\nTransition Probabilities:")
print(transition_probabilities)

# Function to simulate churn behavior using the Markov Model
def simulate_churn(start_state, steps, transition_probabilities):
    state = start_state
    states = [state]
    for _ in range(steps):
        # Choose the next state based on the current state's probabilities
        state = np.random.choice([0, 1], p=transition_probabilities[state])
        states.append(state)
    return states

# Example simulation
initial_state = 0  # Starting with 'Not Churned' (0)
simulation_steps = 10
simulation_result = simulate_churn(initial_state, simulation_steps, transition_probabilities)

print("\nSimulation Result (0 = Not Churned, 1 = Churned):")
print(simulation_result)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Assuming 'transition_matrix' and 'transition_probabilities' have been calculated as in previous steps

# 1. Display the Transition Matrix
print("Transition Matrix:")
print(transition_matrix)

# 2. Display Transition Probabilities
print("\nTransition Probabilities:")
print(transition_probabilities)

# 3. Visualize Transition Probabilities with a Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(transition_probabilities, annot=True, fmt=".2f", cmap="Blues",
            xticklabels=['Not Churned', 'Churned'],
            yticklabels=['Not Churned', 'Churned'])
plt.title("Transition Probability Matrix for Churn")
plt.xlabel("Next State")
plt.ylabel("Current State")
plt.show()

# 4. Summarize churn-related probabilities
churn_rate = transition_probabilities[0, 1]  # Probability of transitioning from Not Churned to Churned
retain_rate = transition_probabilities[0, 0]  # Probability of remaining in Not Churned
print(f"\nChurn Rate (Not Churned -> Churned): {churn_rate:.2f}")
print(f"Retention Rate (Not Churned -> Not Churned): {retain_rate:.2f}")

# 5. Simulate Future Churn Probabilities
def simulate_churn_probabilities(start_state, steps, transition_probabilities):
    state = start_state
    states = [state]
    for _ in range(steps):
        state = np.random.choice([0, 1], p=transition_probabilities[state])
        states.append(state)
    return states

# Simulate churn behavior for 12 months
simulation_steps = 12
initial_state = 0  # Starting with 'Not Churned'
simulation_result = simulate_churn_probabilities(initial_state, simulation_steps, transition_probabilities)

# Count occurrences of churn and non-churn in the simulation results
simulation_counts = pd.Series(simulation_result).value_counts().sort_index()

# 6. Display Simulation Results
plt.figure(figsize=(8, 4))
simulation_counts.plot(kind="bar", color=["blue", "orange"])
plt.title("Simulation of Customer Churn Over Steps")
plt.xlabel("Churn Status (0: Not Churned, 1: Churned)")
plt.ylabel("Number of Customers")
plt.xticks(ticks=[0, 1], labels=["Not Churned", "Churned"])
plt.show()

# Output simulation result details
print("\nSimulation Results (0 = Not Churned, 1 = Churned):")
print(simulation_result)
```
